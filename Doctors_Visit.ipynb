```python
from fastapi import FastAPI, File, UploadFile
import json
import nbformat
import nest_asyncio
import uvicorn

# Patch the event loop to avoid the 'asyncio.run()' error in Jupyter
nest_asyncio.apply()

app = FastAPI()

@app.post("/upload/")
async def upload_notebook(file: UploadFile = File(...)):
    # Read the uploaded file
    content = await file.read()
    
    # Decode the content to string (assuming it's a JSON notebook)
    notebook_content = content.decode("utf-8")
    
    # Load the notebook content as JSON
    try:
        notebook_json = nbformat.reads(notebook_content, as_version=4)
        return {"status": "success", "notebook": notebook_json}
    except Exception as e:
        return {"status": "error", "message": str(e)}

# Run the FastAPI app
if __name__ == "__main__":
    uvicorn.run(app, host="127.0.0.1", port=8000)
```
